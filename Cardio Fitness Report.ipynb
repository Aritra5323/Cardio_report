```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
GoodFitness = pd.read_csv('CardioGoodFitness.csv')
print(GoodFitness.head(10))
print(GoodFitness.shape)
```

```python
#Identifying any missing data
print(GoodFitness.info())
print(GoodFitness.isna().sum())
```

```python
#Description of data
print(GoodFitness.describe(include="all"))
```

```python
for var1 in ["Age", "Education","Fitness","Income",  "Miles", "Usage" ]:
    GoodFitness[var1].hist(figsize= (5,5))
    plt.ylabel("# of observations")
    plt.title(var1)
    plt.show()
```

```python
import seaborn as sns
```

```python
sns.set(style= "whitegrid")
```

```python
A = sns.boxplot(x="Gender", y="Age", data=GoodFitness)
A.set_title("The Relationship between Age and Gender of Consumers", y= 1.03)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)

GoodFitness.hist(by='Gender',column = 'Income',ax=axes)
plt.suptitle("How does Income vary in Relation to Gender?", x=0.5, y=1.05, ha='center', fontsize='large')
fig.text(0.5, -0.05, 'Income (in $)', ha='center')
fig.text(0.01, 0.5, ' # of Observations', va='center', rotation='vertical')
```

```python
#Created a matrix plot to describe the relationship between Gender and Product Type
pd.crosstab(GoodFitness["Product"], GoodFitness["Gender"])
```

```python
sns.countplot(data=GoodFitness, y="Product", hue="Gender").set(title='What Product is Purchased Most by Gender?')
```

Based on the acquired data, the bar grap indicates that the TM195 is the most consumed model with both genders purchasing this product the most as opposed to the TM498 amd TM789 model. This may conclude that the TM195 is the most affordable model for consumers. In additon, the graph shows that females purchased the TM798 significantly less than males. This may mean the design or structure of this product are catered to the male audience.

```python
sns.countplot(data=GoodFitness, y="Usage", hue="Product").set(title="How does a Consumers' Usage Rating Relate to the Product Purchased? [Graph1]")
```

```python
sns.scatterplot(data= GoodFitness, x='Age', y="Income",hue='Product').set(title="How does Income and Age affect a Consumer's Purchase Decision?")
```

Understanding the dynamics of the social class of the average consumer helps companies to understand the purchasing power. This is because consumers in the same social class exhibit similar purchasing behavior. Thus, creating a marketing scheme that targets the average consumer will result in successful sales. Based on the acquired results, the model suggests that the TM798 Product (in Green) is the most expensive product out of the three options. Visually, there is an indication of a steeper slope compared to the other two products (TM195 and TM498), meaning that most consumers between the ages of 23 and 35 with an income over 55000 USD purchased the TM798 model.
